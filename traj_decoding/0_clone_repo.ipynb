```python
from pathlib import Path
import subprocess

REPO_URL = "https://github.com/pearsonlab/trajectory-decoding.git"

# Change this if you want another location
BASE_DIR = Path.cwd()

REPO_DIR = BASE_DIR / "trajectory-decoding"
print("Target repo path:")
print(REPO_DIR.resolve())
```

```python
if REPO_DIR.exists():
    print("Repo already exists. Skip cloning:")
    print(REPO_DIR.resolve())
else:
    subprocess.run(
        ["git", "clone", REPO_URL, str(REPO_DIR)],
        check=True
    )
    print("Cloned to:")
    print(REPO_DIR.resolve())
```

```python
print("After cloning, open Command Prompt and run:")
print()
print(f'cd "{REPO_DIR.resolve()}"')
print("uv sync")
print("uv run python scripts/run_experiment.py")
```
